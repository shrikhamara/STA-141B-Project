```python
# Data Source #1 Yahoo Finance 11/19/2023
import requests
from bs4 import BeautifulSoup

url = "https://finance.yahoo.com/most-active/"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Extract the data you need by inspecting the updated HTML structure
    companies = soup.find_all("tr", class_="simpTblRow")

    for company in companies:
        company_name = company.find("td", class_="Va(m) Ta(start) Px(10px) Fz(s)")
        stock_price = company.find("td", class_="Va(m) Ta(end) Pstart(20px) Fw(600) Fz(s)")

        # Check if the elements exist before extracting text
        if company_name and stock_price:
            company_name = company_name.text.strip()
            stock_price = stock_price.find("fin-streamer").text.strip()

            print(f"Company: {company_name}, Stock Price: {stock_price}")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
# Data Source Yahoo Finance Most Active Stocks: 11/23/23 Stored in CSV file
import requests
from bs4 import BeautifulSoup
import csv

url = "https://finance.yahoo.com/most-active/"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Extract the data you need by inspecting the updated HTML structure
    companies = soup.find_all("tr", class_="simpTblRow")

    # Create a list to store the data
    data = []

    for company in companies:
        company_name = company.find("td", class_="Va(m) Ta(start) Px(10px) Fz(s)")
        stock_price = company.find("td", class_="Va(m) Ta(end) Pstart(20px) Fw(600) Fz(s)")

        # Check if the elements exist before extracting text
        if company_name and stock_price:
            company_name = company_name.text.strip()
            stock_price = stock_price.find("fin-streamer").text.strip()

            # Append data to the list
            data.append([company_name, stock_price])

    # Specify the CSV file path
    csv_file_path = "stock_data.csv"

    # Write the data to the CSV file
    with open(csv_file_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        # Write the header
        writer.writerow(['Company Name', 'Stock Price'])
        # Write the data
        writer.writerows(data)

    print(f"Data has been successfully written to {csv_file_path}")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.offline import plot
from polygon import RESTClient

def fetch_and_plot_stock_data(api_key, ticker, start_date, end_date):
    client = RESTClient(api_key)
    try:
        data_request = client.get_aggs(ticker=ticker,
                                       multiplier=1,
                                       timespan='day',
                                       from_=start_date,
                                       to=end_date)
        price_data = pd.DataFrame(data_request)
        if not price_data.empty:
            price_data['Date'] = price_data['timestamp'].apply(lambda x: pd.to_datetime(x, unit='ms'))
            price_data = price_data.set_index('Date')
            fig = go.Figure(data=[go.Candlestick(x=price_data.index,
                                                 open=price_data['open'],
                                                 high=price_data['high'],
                                                 low=price_data['low'],
                                                 close=price_data['close'])])
            plot(fig, auto_open=True)
        else:
            print("Data is empty. Check the API call parameters.")
    finally:
        client.close()

polygonAPIkey = 'ijCpCwkBDJAslbOwfAc2Ick2iUOomX0B'

# Plot for Tesla
#fetch_and_plot_stock_data(polygonAPIkey, 'TSLA', '2023-01-01', '2023-02-01')
```
